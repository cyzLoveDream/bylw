```python
import xml.dom.minidom 
from xml.dom.minidom import parse
```

```python
t = parse("../data_set/Laptops_Train.xml")
```

```python
dom = parse("../data_set/Laptops_Train.xml")
result = []
dom.hasChildNodes()
for s in dom.getElementsByTagName("sentence"):
    s_id = s.getAttribute("id")
    text = s.getElementsByTagName("text")[0].firstChild.data
    ats = s.getElementsByTagName("aspectTerm")
    terms = []
    polaritys = []
    if ats:
        for i in range(len(ats)):
            terms.append(ats[i].getAttribute("term"))
            polaritys.append(ats[i].getAttribute("polarity"))
    content = {"id":s_id, "text": text, "terms":terms, "polarity":polaritys}
    result.append(content)
```

```python
dom = parse("../data_set/restaurants-trial.xml")
result = []
for s in dom.getElementsByTagName("sentence"):
    s_id = s.getAttribute("id")
    text = s.getElementsByTagName("text")[0].firstChild.data
    ats = s.getElementsByTagName("aspectTerm")
    asc = s.getElementsByTagName("aspectCategory")
    terms = []
    term_polaritys = []
    cates = []
    cate_polaritys = []
    if ats:
        for i in range(len(ats)):
            terms.append(ats[i].getAttribute("term"))
            term_polaritys.append(ats[i].getAttribute("polarity"))
    if asc:
        for i in range(len(asc)):
            cates.append(asc[i].getAttribute("category"))
            cate_polaritys.append(asc[i].getAttribute("polarity"))
    content = {"id": s_id,"text": text,"terms": terms,"term_polarity": term_polaritys,"category":cates,
                        "category_polarity":cate_polaritys}
    result.append(content)
```

```python
#coding=utf-8
import tensorflow as tf
import numpy as np
# 
X = np.random.randn(2, 10, 8)

# example6
X[1,6:] = 0
X_lengths = [10, 6]

cell = tf.contrib.rnn.BasicLSTMCell(num_units=64, state_is_tuple=True)

outputs, last_states = tf.nn.dynamic_rnn(
    cell=cell,
    dtype=tf.float64,
    sequence_length=X_lengths,
    inputs=X)

result = tf.contrib.learn.run_n(
    {"outputs": outputs, "last_states": last_states},
    n=1,
    feed_dict=None)
```
