```python
import json
import pandas as pd
import numpy as np
import warnings
import sys
warnings.filterwarnings("ignore")
```

```python
data = []
with open("../raw_data/DATA_T.txt","r",encoding="utf-8") as rf:
    lines = rf.readlines()
    for l in lines:
        data.append(eval(l))
data1 = []
for d in data:
    data1.append(eval(str(d)))
d_pd = pd.DataFrame(data1)
```

```python
d_pd[["body"]].to_csv("../data_set/w2v_raw_crops.csv",index=False,encoding="utf-8")
```

```python
d = []
with open("E:\\bylw\\raw_data\\DATA_T_RESULT.txt","r",encoding="utf-8") as fs:
    lines = fs.readlines()
    for line in lines:
        d.append(eval(str(line)))
```

```python
pd_data = pd.DataFrame()
for i in d:
    i_pd = pd.DataFrame(i.get("entities"))
    pd_data = pd.concat([pd_data, i_pd],axis=0,ignore_index=True)
```

```python
pd_data.head()
```

```python
pd_data.groupby("eventLevel").count()
```

```python
pd_data["digest"]
```

```python
ts = pd_data[["name","digest","eventLevel"]]
```

```python
ts.shape
```

```python
ts = ts[ts["digest"] != ""]
```

```python
ts.shape
```

```python
ts["eventLevel"] = ts["eventLevel"].apply(lambda x: 0 if x == "正向" else 1 if x == "中性" else 2)
```

```python
ts.head()
```

```python
def tihuan(s):
    s1 = s[0]
    if type(s1) is not str:
        return ""
    s2 = s[1]
    return s2.replace(s1, "$T$")
```

```python
ts["digest1"] = ts[["name","digest"]].apply(tihuan,axis=1)
ts.head()
```

```python
ts.ix[:5,:].values
```

```python
ts = ts[["name","digest1","eventLevel"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
trains, tests = train_test_split(ts, random_state = 42, train_size=0.8)
```

```python
trains.shape
```

```python
train, evals = train_test_split(trains, random_state = 42, train_size=0.8)
```

```python
tests.shape
```

```python
with open("E:\\bylw\\data_set\\train.csv","w+",encoding="utf-8") as wf:
    for i in train.values.tolist():
        wf.write(i[1] + "\n" + i[0] + "\n" + str(i[2]) + "\n")
```

```python
with open("E:\\bylw\\data_set\\valid.csv","w+",encoding="utf-8") as wf:
    for i in evals.values.tolist():
        wf.write(i[1] + "\n" + i[0] + "\n" + str(i[2]) + "\n")
```

```python
with open("E:\\bylw\\data_set\\test.csv","w+",encoding="utf-8") as wf:
    for i in tests.values.tolist():
        wf.write(i[1] + "\n" + i[0] + "\n" + str(i[2]) + "\n")
```
